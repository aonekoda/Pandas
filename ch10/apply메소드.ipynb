# apply 메소드

**함수의 적용과 mapping**  
각 row 나 컬럼의 배열에 함수를 적용하는 것이 가능하다. dataframe의 apply 메소드를 통해 수행 할 수 있다.

```python
import pandas as pd
import numpy as np
```

```python
frame = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
frame
```

```python
f = lambda x: x.max() - x.min()

frame.apply(f)
```

```python
frame.apply(f, axis='columns')
```

```python
def f(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])

frame.apply(f)
```

### 그룹 연산과 apply 메소드

```python
tips = pd.read_csv('../data/tips.csv')
tips.head()
```

```python
# 전체 금액 대비 팁의 비율
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

상위 5위 `top_pct` 값을 출력하는 함수를 작성한다.

```python
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]

top(tips)
```

`smoker` 그룹에 대해서 이 함수를 `apply` 하게 되면 다음과 같은 결과를 얻을 수 있다.

```python
tips.groupby('smoker').apply(top)
```

```python
# 다음과 같이 적용하는 것도 가능하다.
# 흡연석별, 요일별 total_bill 의 최대값(top=1)을 계산한다
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```
